```python
def find_average(numbers):
    
    total = 0
    count = 0

    for num in numbers:
        total += num
        count += 1

    if count == 0:
        return 0

    average = total / count
    return average



nums = [10, 20, 30, 40, 50]

avg = find_average(nums)

print("Numbers:", nums)
print("Average value is:", avg)
```
