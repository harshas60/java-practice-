Program to input various values from user and print their type

```python
n_int = int(input("Enter an integer: "))
n_str = input("Enter a string: ")
n_float = float(input("Enter a float: "))
n_bool = bool(input("Enter a boolean (true/false): ").capitalize())

print("Integer:", n_int, type(n_int))
print("String:", n_str, type(n_str))
print("Float:", n_float, type(n_float))
print("Boolean:", n_bool, type(n_bool))
```

Program to input month and year from user and print the month calendar

```python
import calendar

m = int(input("Input the calendar month: "))
y = int(input("Year: "))
print("Calendar of:")
print(calendar.month(y, m))
```
