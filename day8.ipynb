```python
def life_in_week(age):
    years_left = (90 - age) *52
    print(f"You have {years_left} weeks left.")
life_in_week(52)   
    
```

```python
def greet():
    print("Hello")
    print("How do you do?")
    print("Isn't the weather nice today?")
greet()    
```

```python
# function with doing something which is called side effects
def greet_with_name(name):
    print(f"Hello {name}")
    print(f"How do you do {name}?")
    print(f"Isn't the weather nice today, {name}?") 
greet_with_name("Harsha")    
```

```python
#  to use function with more than one input
def greet_with(name, location):
    print(f"Hello {name}")
    print(f"What is it like in {location}?")
greet_with("Harsha", "India")
greet_with(location="India", name="Harsha")
```

```python
def calculate_love_score(name1, name2):
    combined_names = (name1 + name2).lower()

    true_count = (
        combined_names.count('t') +
        combined_names.count('r') +
        combined_names.count('u') +
        combined_names.count('e')
    )

    love_count = (
        combined_names.count('l') +
        combined_names.count('o') +
        combined_names.count('v') +
        combined_names.count('e')
    )

    print(str(true_count) + str(love_count))



calculate_love_score("Kanye West", "Kim Kardashian")
```
