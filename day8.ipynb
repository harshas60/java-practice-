```python
def life_in_week(age):
    years_left = (90 - age) *52
    print(f"You have {years_left} weeks left.")
life_in_week(52)   
    
```

```python
def greet():
    print("Hello")
    print("How do you do?")
    print("Isn't the weather nice today?")
greet()    
```

```python
# function with doing something which is called side effects
def greet_with_name(name):
    print(f"Hello {name}")
    print(f"How do you do {name}?")
    print(f"Isn't the weather nice today, {name}?") 
greet_with_name("Harsha")    
```
